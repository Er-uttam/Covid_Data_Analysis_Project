```python
covid_df.drop(['Sno','Time','ConfirmedIndianNational','ConfirmedForeignNational'],inplace=True, axis=1)
```

```python
covid_df.head()
```

```python
covid_df['Date'] = pd.to_datetime(covid_df['Date'], format ='%Y-%m-%d')
```

```python
covid_df.head()
```

```python
# Active cases
covid_df['Active_cases'] = covid_df['Confirmed']-(covid_df['Cured']+covid_df['Deaths'])
covid_df.head()
```

```python
statewise = pd.pivot_table(covid_df , values=['Confirmed','Deaths','Cured'],index="State/UnionTerritory",aggfunc=max)
```

```python
statewise["Recovery Rate"] = statewise['Cured']*100/statewise['Confirmed']
```

```python
statewise["Mortality Rate"] = statewise['Deaths']*100/statewise['Confirmed']
```

```python
statewise = statewise.sort_values(by='Confirmed', ascending = False)
```

```python
statewise.style.background_gradient(cmap = 'cubehelix') # cmap i.e colors map
```

```python
# top 10 active cases states
top_10_active_cases = covid_df.groupby(by='State/UnionTerritory').max()[['Active_cases','Date']].sort_values(by='Active_cases',ascending=False).reset_index()

fig = plt.figure(figsize =(16,9))
plt.title("Top 10 States with most active cases in India")
ax = sns.barplot(data = top_10_active_cases.iloc[:10] , y="Active_cases" , x='State/UnionTerritory' , linewidth = 2 ,edgecolor='red',palette='bright')

plt.xlabel("States")
plt.ylabel("Total Active cases")
plt.show()
```

```python
# Top states with highest deaths
top_10_deaths = covid_df.groupby(by='State/UnionTerritory').max()[['Deaths','Date']].sort_values(by='Deaths',ascending=False).reset_index()

fig = plt.figure(figsize =(18,8))
plt.title("Top 10 States with most Deaths" , size=25)
ax = sns.barplot(data = top_10_deaths.iloc[:12] , y="Deaths" , x='State/UnionTerritory' ,linewidth =2,edgecolor='black',palette='bright')

plt.xlabel("States")
plt.ylabel("Total Deaths cases")
plt.show()
```

```python
# Growth trend
fig = plt.figure(figsize = (12,6))
ax = sns.lineplot(data =covid_df[covid_df['State/UnionTerritory'].isin(['Maharashtra','Karnataka','Kerala','Tamil Nadu','Uttar Pradesh'])],x='Date',y='Active_cases',hue='State/UnionTerritory')

ax.set_title("Top 5 Affected states in India", size=16)
plt.show() # Assuming you want to display the plot
```

```python
# Remove rows where state = india
vaccine = vaccine_df[vaccine_df.State!= 'India']
vaccine
```

```python
vaccine.rename(columns={"Total Individuals Vaccinated":"Total"},inplace=True)
```

```python
vaccine.head()
```

```python
# most vaccinated state
max_vac = vaccine.groupby('State')['Total'].sum().to_frame('Total')
max_vac = max_vac.sort_values('Total',ascending = False)[:5]
max_vac
```

```python
fig = plt.figure(figsize=(10,5))
plt.title("Top 5 Vaccinated states in India", size =20)
x = sns.barplot(data = max_vac.iloc[:10],y=max_vac.Total, x = max_vac.index,linewidth = 2 , edgecolor ='black', palette='bright') # Add palette='bright' here to plot the bar of different color
plt.xlabel("States")
plt.ylabel("Vaccination")
plt.show()
```

```python
# Least vaccinated state
min_vac = vaccine.groupby('State')['Total'].sum().to_frame('Total')
min_vac = max_vac.sort_values('Total',ascending = True)[:5]
min_vac
```

```python
fig = plt.figure(figsize=(15,7))
plt.title("Least Vaccinated states in India", size =20)
x = sns.barplot(data = min_vac.iloc[:10],y=min_vac.Total, x = min_vac.index,linewidth = 2 , edgecolor ='black', palette='bright') # Add palette='bright' here
plt.xlabel("States")
plt.ylabel("Vaccination")
# plt.xticks(rotation=45)
plt.show()
```
